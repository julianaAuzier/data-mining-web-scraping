```python
import re
lista_pesquisa = ['sintomas','COVID-19']
```

```python
#lista de termos para busca
lista_pesquisa = ['sintomas','COVID-19']
```

```python
#extraindo somente o texto da pgina
texto = soup.get_text()
texto
```

```python
for item in lista_pesquisa:
    print('buscando por "%s" no texto ' %(item))
    if re.search(item, texto):
        print('\n')
        print('Palavra encontrada')
        print('\n')
    else:
        print('\n')
        print('Palavra no encontrada')
        print('\n')
```

```python
#usando a funao split
termo_de_quebra = '@'
frase = 'o dominio do email de alguem com o e-mail : nomedouser@gmail.com'
re.split(termo_de_quebra, frase)
```

```python
#expressoes regulares - buscando padres
def encontra_padrao(lista_padroes, texto_):
    for item in lista_padroes:
        print('procurando por "%s" no texto...\n' %(item))
        print(re.findall(item, texto_))
        print('\n')
```

```python
#o site fala sobre quais doenas?
lista_p = ['doena...............']
encontra_padrao(lista_p, texto)
```

```python
#o site fala sobre gravidez?
lista_p = ['gesta....{1}']
encontra_padrao(lista_p, texto)
```

```python
#o site fala sobre remdios caseiros?
lista_p = ['rem......cas.....']
encontra_padrao(lista_p, texto)
```
